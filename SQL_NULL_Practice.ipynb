# **SQL NULL Values**

*   Update Statement
*   Delete Statement

*   Alter Table
*   Drop Column





```python
create database customers;
show databases;
use customers;
```

Screenshot (38).png

```python
#create table 
create table customer_info(
id integer auto_increment,
first_name varchar(25),
last_name varchar(25),
salary integer,
primary key(id)
);
```

```python
#insert records in table 
insert into customer_info(first_name,last_name,salary)
values
('sai','shinde',20000),
('sambhaji','sonwane',30000),
('suraj','Gaikwad',35000),
('prasad','nalwade',40000),
('umesh','adhari',NULL);
```

```python
select * from customer_info; # it is showing table
```

# **O/P:-**
Screenshot (39).png

```python
#how to check where is null value in table
select * from customer_info where salary is null;
```

# **O/P:-**
Screenshot (40).png

```python
#it is showing not present in null value 
select * from customer_info where salary is not null;
```

Screenshot (41).png

```python
#sql update statement to replace null values
update customer_info set salary=50000 where id=5;

select * from customer_info;
```
